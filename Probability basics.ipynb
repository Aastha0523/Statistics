```python
#Random Variable

import numpy as np

# Simulate a discrete RV: dice roll
dice = np.random.randint(1, 7, 10)
print("Dice rolls:", dice)

# Simulate a continuous RV: normal distribution
heights = np.random.normal(loc=170, scale=10, size=10)
print("Heights:", np.round(heights, 1))
```

```python
#Probability Distribution

import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# --- Discrete: Binomial ---
n, p = 10, 0.5
x = np.arange(0, n+1)
pmf_vals = binom.pmf(x, n, p)   # PMF
cdf_vals = binom.cdf(x, n, p)   # CDF

plt.bar(x, pmf_vals, alpha=0.6, label="PMF")
plt.plot(x, cdf_vals, 'o-', color="red", label="CDF")
plt.title("Binomial(10, 0.5)")
plt.legend()
plt.show()

# --- Continuous: Normal ---
x = np.linspace(-4, 4, 100)
pdf_vals = norm.pdf(x, 0, 1)   # PDF
cdf_vals = norm.cdf(x, 0, 1)   # CDF

plt.plot(x, pdf_vals, label="PDF")
plt.plot(x, cdf_vals, label="CDF")
plt.title("Normal(0,1)")
plt.legend()
plt.show()
```
